```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
#1 is buy for update type
```

```python
df = pd.read_csv('C:\\Users\\Jason\\Documents\\Python\\Stock Market Prediction\\Crypto\\exch_gdax_btcusd_snapshot_20171004.csv', nrows=100000)
```

```python
df.head(3)
```

```python
#tdt=tdt.join(df["trade_px"])
df['trade_px']=df[["b1", "a1"]].mean(axis=1)

df.head(110)
```

```python

df['sign_vol']=df['trade_volume'].loc[df['update_type'] == 1]
df['sign_vol']=df['sign_vol'].combine_first(-df['trade_volume'].loc[df['update_type'] == 2])
df.head(2)
```

```python
df['aqsum5']= df['aq1']+df['aq2']+df['aq3']+df['aq4']+df['aq5']
df['bqsum5']= df['bq1']+df['bq2']+df['bq3']+df['aq4']+df['aq5']
df['abq']= df['aq1']-df['bq1']
df['abq5']= df['aqsum5']-df['bqsum5']
```

```python
tdt=pd.to_datetime(df.trades_date_time).to_frame()


tdt['sign_vol']=df['sign_vol']
tdt.index = pd.to_datetime(df['trades_date_time'])
tdt=tdt.drop(tdt.index[:5])
tdt = tdt.resample('s').mean()
tdt=tdt.interpolate(method='linear')
tdt.index.rename('order_date_time', inplace=True)
tdt['sign_vol']=tdt['sign_vol'].rolling(window= 5, center=False).mean()
tdt.head(2)
```

```python
#date_time=pd.to_datetime(df.trades_date_time).to_frame()
odt=pd.to_datetime(df.order_date_time).to_frame()
odt["Spread"] = df["a1"]-df["b1"]
odt['trade_px']=df['trade_px']

odt["a1"] = df["a1"]
odt["b1"] = df["b1"]
odt["aq1"] = df["aq1"]
odt["bq1"] = df["bq1"]
odt["aqsum5"] = df["aqsum5"]
odt["bqsum5"] = df["bqsum5"]
odt['abq']=df['abq']
odt['abq5']=df['abq5']
odt['abratio']= df['aq1']/(df['aq1']+df['bq1'])
odt['id']=df['id']

#.rolling(window= 1, center=False).mean()
```

```python

#date_time_vol=date_time.join(df["trade_volume"])
#sign=df["trade_volume"].subtract(df["aq1"])
#sign=date_time_vol["sign"].add(0.0001)
#sign[sign >= 0] = 1
#sign[sign < 0] = -1
odt["trade_volume"]=df["trade_volume"]
#.multiply(sign)
```

```python
odt=odt.drop(odt.index[:5])
odt.head(2)
```

```python
odt.index = pd.to_datetime(odt['order_date_time'])
odt = odt.resample('s').mean()
odt=odt.interpolate(method='linear')

odt.head(2)
```

```python
tdt.head(2)
```

```python
#odt['abqdt']=pd.rolling_mean((odt['aq1']-odt['bq1']), 15)
odt['abqdt']=(odt['abq']-(odt['abq'].shift(-1))).rolling(window= 2, center=False).mean()

odt['abqdt^2']=(odt['abqdt']-(odt['abqdt'].shift(-1)))

#odt['abqdt']=(odt['aq1']-odt['bq1']).rolling(window= 15, center=False).mean()
odt['rollingRatio']=odt['abratio'].rolling(window= 5, center=False).mean()
odt['ratio5']=odt['aqsum5']/(odt['aqsum5']+odt['bqsum5'])
odt['ratio5dt']=(odt['ratio5']-(odt['ratio5'].shift(-1))).rolling(window= 2, center=False).mean()
odt=pd.concat([odt, tdt], axis=1)
odt.head(12)
```

```python
plt.figure(figsize=(25,7))
plt.plot(odt['trade_px'], label = 'trade_px')
plt.legend(loc='upper left')
plt.show()
```

```python
odt['diff']=odt['trade_px'].diff(periods=1).round(5)
```

```python
odt['norepeat']=odt['diff'].loc[odt['diff'].shift(-1) != 0]
diff = odt[odt['diff'] != 0]
nans = np.isnan(diff)
odt.head(120)
```

```python
sliced = odt[odt['diff'] != 0]
start = odt[odt['norepeat'] == 0]
sliced=sliced.iloc[::5]
start=pd.concat([sliced,start])
start.head(10)
```
